# Anndata 구축

```python
# df1: TMM X, CPM X 
# df2: TMM X, CPM O 
# df3: TMM O, CPM O 
# df4: TMM O, log(CPM)
# df5: TMM O, log(CPM)

df1 = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_count_matrix.txt')
df2 = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix_log_false_No_TNM.txt')
df3 = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix_log_false.txt')
df4 = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix.txt')
df5 = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix_corrected.txt')
```

```python
counts = df1[meta.index].T
adata = ad.AnnData(X=df1[meta.index].T, obs=meta)

adata.layers['df1'] = df1[meta.index].T
adata.layers['df2'] = df2[meta.index].T
adata.layers['df3'] = df3[meta.index].T
adata.layers['df4'] = df4[meta.index].T
adata.layers['df5'] = df5[meta.index].T

adata
```

```python
my_layers = ['df1','df2','df3','df4','df5']

for my_layer in my_layers:
    # PCA -----------------------------
    adata.X = adata.layers[my_layer]
    sc.tl.pca(adata)
    sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
    
    # Boxplot -------------------------
    df = pd.DataFrame(adata.layers[my_layer].T, index=adata.var_names, columns=adata.obs_names)
    plt.figure(figsize=(20,4))
    df.boxplot(rot=90)
    
    plt.grid(False)
    plt.title('Boxplot of Gene Expression')
    plt.xlabel('Samples')
    plt.ylabel('Counts')
    plt.show()
```

```python
adata = DeseqDataSet(counts=df1[meta.index].T,
                     metadata=pd.DataFrame(adata.obs),
                     design_factors="Condition")
adata
```

```python
adata.layers['df1'] = df1[meta.index].T
adata.layers['df2'] = df2[meta.index].T
adata.layers['df3'] = df3[meta.index].T
adata.layers['df4'] = df4[meta.index].T
adata.layers['df5'] = df5[meta.index].T
```

```python
adata.deseq2()
```

```python
adata
```

```python
# cpm
adata.X = adata.layers['df1']
lib_size = np.sum(adata.X, axis=1)
lib_size = lib_size[:, np.newaxis]
cpm_values = (adata.X / lib_size) * 1e6
cpm_values[0:4,  0:4]
adata.layers['my_cpm'] = cpm_values

# my_normed_cpm
adata.X = adata.layers['normed_counts']
lib_size = np.sum(adata.X, axis=1)
lib_size = lib_size[:, np.newaxis]
cpm_values = (adata.X / lib_size) * 1e6
cpm_values[0:4,  0:4]
adata.layers['my_normed_cpm'] = cpm_values

# my_normed_log
adata.layers['my_normed_log'] = np.log1p(adata.layers['normed_counts'])

# my_normed_cpm_log
adata.layers['my_normed_cpm_log'] = np.log1p(adata.layers['my_normed_cpm'])
```

```python
my_layers = ['my_cpm','normed_counts','my_normed_cpm','my_normed_log','my_normed_cpm_log']

for my_layer in my_layers:
    print(my_layer)
    
    # PCA -----------------------------
    adata.X = adata.layers[my_layer]
    sc.tl.pca(adata)
    sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
    
    # Boxplot -------------------------
    df = pd.DataFrame(adata.layers[my_layer].T, index=adata.var_names, columns=adata.obs_names)
    plt.figure(figsize=(20,4))
    df.boxplot(rot=90)
    
    plt.grid(False)
    plt.title('Boxplot of Gene Expression')
    plt.xlabel('Samples')
    plt.ylabel('Counts')
    plt.show()
```
