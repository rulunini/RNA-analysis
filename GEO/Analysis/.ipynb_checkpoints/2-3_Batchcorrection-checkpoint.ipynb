```python
import pickle
import os

import pandas as pd
import scanpy as sc
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pydeseq2
from pydeseq2.dds import DeseqDataSet
from pydeseq2.default_inference import DefaultInference
from pydeseq2.ds import DeseqStats
from pydeseq2.utils import load_example_data

from sklearn.preprocessing import FunctionTransformer
from scipy import stats
import anndata as ad
```

# Load data

```python
os.chdir('/Users/soheepark/03-GEO근감소/Data/')
```

```python
# Load GSE167186 data
meta16 = pd.read_csv('GSE167186_metadata.csv', sep='\t', index_col=[0])
counts16 = pd.read_csv('GSE167186_counts.csv', sep='\t', index_col=[0], low_memory=False)
meta16_2 = pd.read_csv('/Users/soheepark/03-GEO근감소/Data/GSE167186_metadata2.csv', index_col=0)

# Load GSE111016 data
meta11 = pd.read_csv('./GSE111016_metadata.csv', sep='\t', index_col=[0])
counts11 = pd.read_csv('./GSE111016_counts.csv', sep='\t', index_col=[0], low_memory=False)
```

```python
meta16_2['Temp4'] = meta16_2.index
meta16_2.index = meta16_2['Temp3']
```

```python
meta16.index = counts16.index.map(meta16_2['Temp4'])
meta16['Condition'] = meta16['Condition'].map({'Sarcopenia':'OS','Old Healthy':'OH', 'Young Healthy':'YH'})
meta16['GEO'] = 'GSE167186'
meta16['Batch'] = 0
```

```python
meta16_2.head()
```

```python
meta11['GEO'] = 'GSE111016'
meta11['Batch'] = 1
```

```python
meta = pd.concat([meta11, meta16], axis=0)
meta.index = pd.Series(meta.index).apply(lambda x:x.replace(' ', '.'))
meta.head()
```

# Anndata 구축

```python
count_df = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_count_matrix.txt')
cpm_df = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix.txt')
cpm_cor_df = pd.read_table('/Users/soheepark/03-GEO근감소/Data/merged_cpm_matrix_corrected.txt')
```

```python
counts = count_df[meta.index].T
adata = ad.AnnData(X=counts, obs=meta)
adata.layers['raw'] = adata.X
adata.layers['cpm'] = cpm_df[meta.index].T
adata.layers['cpm_cor'] = cpm_cor_df[meta.index].T
adata
```

```python
adata.X = adata.layers['raw']
sc.tl.pca(adata)
sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
```

```python
adata.X = adata.layers['cpm']
sc.tl.pca(adata)
sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
```

```python
adata.X = adata.layers['cpm_cor']
sc.tl.pca(adata)
sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
```

# 아래는 CPM 계산 

```python
adata.X = adata.layers['raw']
dds = DeseqDataSet(counts=adata.X,
                   metadata=pd.DataFrame(adata.obs),
                   design_factors="Condition")
dds
```

```python
dds.deseq2()
```

```python
dds.layers['normed_counts']
```

```python
# PCA plotting
dds.X = dds.layers['normed_counts']
sc.tl.pca(dds)
sc.pl.pca(dds, color='GEO', size=200, alpha=0.3)
```

```python
# Self calculation of CPM by raw
adata = dds.copy()

lib_size = np.sum(adata.X, axis=1)
lib_size = lib_size[:, np.newaxis]

cpm_values = (adata.X / lib_size) * 1e6
cpm_values[ 0:4,  0:4]
adata.layers['normed_counts_cpm'] = cpm_values

adata.X = adata.layers['normed_counts_cpm']
sc.tl.pca(adata)
sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
```

```python
# Self calculation of CPM by normed_counts
adata = dds.copy()

adata.X = adata.layers['normed_counts']
lib_size = np.sum(adata.X, axis=1)
lib_size = lib_size[:, np.newaxis]

cpm_values = (adata.X / lib_size) * 1e6
cpm_values[ 0:4,  0:4]
adata.layers['normed_counts_cpm'] = cpm_values

adata.X = adata.layers['normed_counts_cpm']
sc.tl.pca(adata)
sc.pl.pca(adata, color='GEO', size=200, alpha=0.3)
```
